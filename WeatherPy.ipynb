## Weather analysis

* Maximum temperatures get higher the closer to the ecuator we are, even considering the different seasons for the northern and souther hemispheres.
* Humidity seems to be higher around the ecuator, with most of the cities over 50% of humidity.
* Wind speed seems to be higher when cities are farther from the ecuator. 
* Cloudiness seems to be independent from the city's latitude.
* High temperatures, high humidity and low wind speed can help understand why the weather gets hotter in cities near the ecuator compared to higher latitudes cities.

```python
import pandas as pd
from citipy import citipy
import requests
import matplotlib.pyplot as plt
from config import (api_key,g_key)
import numpy as np
from numpy import random
```

```python
# Generate list of random coordinates
cities_lon = np.random.uniform(-180,180,2000)
cities_lat = np.random.uniform(-90,90,2000)
cities_coord = list(zip(cities_lat,cities_lon))
len(cities_coord)
```

```python
# Get nearby cities names
city_data = []

for coordinates in cities_coord: 
    lat,long = coordinates
    city = citipy.nearest_city(lat,long).city_name
    if city not in city_data:
        city_data.append(city)
len(city_data)
```

```python
# Get weather data
url = 'https://api.openweathermap.org/data/2.5/weather?'
weather = {}
print('Begining data retrieval')
print('------------------------------------------------')
n = 0  
m = 0
for city in city_data:
    city_url = url + "q=" + city+'&units=metric'+'&'+ "appid=" + api_key 
    weather[city] = requests.get(city_url).json()
       
    if weather[city]['cod']!='404':
        print(f'Processing record {m} | {city}: ')
        print(f'{city_url}')
        n = n+1 
        m = m+1
    else:
        weather.pop(city)
        print(f'Processing record {m}| {city}: ')
        print('Not found, skipping city')
        m = m+1

    if n >= 500:
        break
```

```python
# Retrieve max temp and coordinate of cities
temps = {}
coord_lon = {}
coord_lat = {}

for city in weather:
    temps[city] = weather[city]['main']['temp_max']
    coord_lon[city] = weather[city]['coord']['lon']
    coord_lat[city] = weather[city]['coord']['lat']

# Generate dataframes
temps_df = pd.DataFrame.from_dict(temps,orient='index')
temps_df = temps_df.rename(columns={0:'Max Temp'})

coord_lon_df = pd.DataFrame.from_dict(coord_lon,orient='index')
coord_lon_df = coord_lon_df.rename(columns={0:'Lon'})

coord_lat_df = pd.DataFrame.from_dict(coord_lat,orient='index')
coord_lat_df = coord_lat_df.rename(columns={0:'Lat'})

# Merge dataframes
temp_lon_df = pd.merge(temps_df,coord_lon_df, left_index=True,right_index=True)
temp_coord_df = pd.merge(temp_lon_df,coord_lat_df, left_index=True,right_index=True)

temp_coord_df.head()
```

```python
# Retrieve humidity, cloudiness and wind speed
humid = {}
cloud = {}
wind = {}

for city in weather:
    humid[city] = weather[city]['main']['humidity']
    cloud[city] = weather[city]['clouds']['all']
    wind[city] = weather[city]['wind']['speed']

# Generate dataframes
humid_df = pd.DataFrame.from_dict(humid,orient='index')
humid_df = humid_df.rename(columns={0:'% Humidity'})

cloud_df = pd.DataFrame.from_dict(cloud,orient='index')
cloud_df = cloud_df.rename(columns={0:'% Cloudiness'})

wind_df = pd.DataFrame.from_dict(wind,orient='index')
wind_df = wind_df.rename(columns={0:'Wind speed'})

# Merge dataframes
humid_cloud_df = pd.merge(humid_df,cloud_df, left_index=True,right_index=True)
hum_cl_win_df = pd.merge(humid_cloud_df,wind_df, left_index=True,right_index=True)
weather_df = pd.merge(temp_coord_df,hum_cl_win_df, left_index=True,right_index=True)

# Save cvs file
weather_df.to_csv('Results/weather_data.csv')
weather_df.head()
```

```python
# Temperature plot

plt.scatter(weather_df['Lat'], weather_df['Max Temp'])
plt.xlabel('Latitude')
plt.ylabel('Max Temp (C)')
plt.title("Max temp vs City latitude")
plt.xlim(-90,90)
plt.axvline(0)

plt.savefig('Results/temp_plot.png')
```

```python
# Humidity plot

plt.scatter(weather_df['Lat'], weather_df['% Humidity'])
plt.xlabel('Latitude')
plt.ylabel('% Humidity')
plt.title("Humidity vs City latitude")
plt.xlim(-90,90)
plt.axvline(0)

plt.savefig('Results/hum_plot.png')
```

```python
# Cloudiness plot

plt.scatter(weather_df['Lat'], weather_df['% Cloudiness'])
plt.xlabel('Latitude')
plt.ylabel('% Cloudiness')
plt.title("Cloudiness vs City latitude")
plt.xlim(-90,90)
plt.axvline(0)

plt.savefig('Results/cloud_plot.png')
```

```python
# Wind speed plot

plt.scatter(weather_df['Lat'],weather_df['Wind speed'])
plt.xlabel('Latitude')
plt.ylabel('Wind speed (mph)')
plt.title("Wind speed vs City latitude")
plt.xlim(-90,90)
plt.axvline(0)

plt.savefig('Results/wind_plot.png')
```
